## Calculate bidAskSpreadPct
* bidAskSpreadPct = (askPrice-bigPrice)/idealPrice
* Perform descriptive statistics. Does the data make sense?
  * Negative spreads don't make sense. Trim at 0.0 (replace negative with 0)
* Plot the conditional distributions of bidAskSpreadPct during after-hours and regular trading hours (93000-160000)
  * Use openTime, ignore exitTime.
  * Clearly different distributions, so we need a different strategy per partition.
  * Trim the maximum to be 99th percentile for each partion, but don't override bidAskSpreadPct. Instead, call the new variable bidAskSpreadPctTrim. This way we have a backup of the original.
* Recalculate the correlations to 'entrySlipPct'.
  * How good is this new feature?
  * Does trimming help?
  * Why did the bidPrice or askPrice or idealPrice alone not work well?

````python
df['bidAskSpreadPct'] = (df['askPrice'] - df['bidPrice']) / df['idealPrice']
````

## Adjust the level of trimming

* Search different quantiles to bidAskSpreadPct to maximize the correlation to entrySlipPct
* Eg you could use a grid search
`np.arange(0.9,1.0,0.01)`




````python
quantiles = np.arange(0.9, 1.0, 0.01)
correlations = []
df2 = df
for q in quantiles:
    # Trim bidAskSpreadPct at the current quantile
    temp = df2
    temp['bidAskSpreadPctTrim'] = df2['bidAskSpreadPct'].clip(upper=np.percentile(df2['bidAskSpreadPct'], q * 100))
    # Calculate the correlation
    corr = np.corrcoef(temp['bidAskSpreadPctTrim'], df2['entrySlipPct'])[0, 1]
    correlations.append(corr)
# Find the index of the maximum correlation
max_corr_index = np.argmin(correlations)
max_corr_quantile = quantiles[max_corr_index]

# Trim bidAskSpreadPct on df at the quantile that gives the maximum corr
df['bidAskSpreadPctTrim'] = df['bidAskSpreadPctTrim'].clip(upper=np.percentile(df['bidAskSpreadPctTrim'], max_corr_quantile * 100))
````
